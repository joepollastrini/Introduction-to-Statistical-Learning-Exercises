10) This exercise involves the Boston housing data set.

a) Load data.  How many rows and columns are there and what does each represent?

```python
boston = pd.read_csv(os.path.join(os.getcwd(), 'Boston.csv')
                     , delimiter=r"\s+"
                     , names=['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax'
                              , 'ptratio', 'b', 'lstat', 'medv'])
boston.info()
```

****Answer:****  There are 506 observations and 14 parameters.  The following was found online for each variable meaning:
* crim - per capita crime rate by town
* zn - proportion of residential land zoned for lots over 25000 sq. ft.
* indus - proportion of non-retail business acres per town
* chas - Charles River dummy variable (1 if tract bounds river)
* nox - nitric oxides concentration (parts per 10 million)
* rm - average number of rooms per dwelling
* age - proportion of owner-occupied units built prior to 1940
* dis - weighted distances to five Boston employment centres
* rad - index of accessibility to radial highways
* tax - full value property tax rate per USD 10000
* ptratio - pupil to teacher ratio
* b - 1000(B - 0.63)^2 where B is the proportion of blacks by town
* lstat - percentage of lower status of the population
* medv - median value of owner occupied homes in USD 1000

c) Are there any predictors associated with per capita crime rate?

```python
pd.plotting.scatter_matrix(boston, figsize=(15,10));
```

```python
fig = plt.figure(figsize=(12, 15), dpi= 80, facecolor='w', edgecolor='k')
for i, c in enumerate(boston.columns[1:]):
    fig.add_subplot(len(boston.columns[1:])/3 + 1, 3, i+1)
    sns.regplot(x=boston[c], y=boston['crim']);
```

```python
len(boston['rad'].unique())
```

****Answer:****
* ****zn**** seems to be indicative of crime rate.  Specifically zn != 0 implies no crime rates.
* a specific ****indus**** value implies crime rates > 0
* ****chas**** - being along the charles river implies a low crime rate
* a high ****rad**** value implies crime rates > 0
* ****age**** - a larger proportion of older homes implies a larger crime rate
* ****medv**** - higher median home value implies lower crime rates

* ****rad**** only has 9 unique values, so could probably be treated as a qualitative variable with grouping
